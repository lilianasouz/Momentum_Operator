```python
from  qiskit import*
circuit = QuantumCircuit(7,7)
from math import pi
```

```python
circuit.barrier()
circuit.p(pi/8,5)
circuit.p(pi/16,4)
circuit.p(pi/32,3)
circuit.p(pi/64,2)
circuit.p(pi/128,1)
circuit.p(pi/256,0)

circuit.cx(5,6)
circuit.cx(4,6)
circuit.p(2*pi,6)
circuit.cx(4,6)
circuit.cx(5,6)


circuit.cx(5,6)
circuit.cx(3,6)
circuit.p(pi,6)
circuit.cx(3,6)
circuit.cx(5,6)

circuit.cx(5,6)
circuit.cx(2,6)
circuit.p(pi/2,6)
circuit.cx(2,6)
circuit.cx(5,6)

circuit.cx(5,6)
circuit.cx(1,6)
circuit.p(pi/4,6)
circuit.cx(1,6)
circuit.cx(5,6)

circuit.cx(5,6)
circuit.cx(0,6)
circuit.p(pi/8,6)
circuit.cx(0,6)
circuit.cx(5,6)

circuit.cx(4,6)
circuit.cx(3,6)
circuit.p(pi/2,6)
circuit.cx(3,6)
circuit.cx(4,6)

circuit.cx(4,6)
circuit.cx(2,6)
circuit.p(pi/4,6)
circuit.cx(2,6)
circuit.cx(4,6)

circuit.cx(4,6)
circuit.cx(1,6)
circuit.p(pi/8,6)
circuit.cx(1,6)
circuit.cx(4,6)

circuit.cx(4,6)
circuit.cx(0,6)
circuit.p(pi/16,6)
circuit.cx(0,6)
circuit.cx(4,6)

circuit.cx(3,6)
circuit.cx(2,6)
circuit.p(pi/8,6)
circuit.cx(2,6)
circuit.cx(3,6)

circuit.cx(3,6)
circuit.cx(1,6)
circuit.p(pi/16,6)
circuit.cx(1,6)
circuit.cx(3,6)

circuit.cx(3,6)
circuit.cx(0,6)
circuit.p(pi/32,6)
circuit.cx(0,6)
circuit.cx(3,6)

circuit.cx(2,6)
circuit.cx(1,6)
circuit.p(pi/32,6)
circuit.cx(1,6)
circuit.cx(2,6)

circuit.cx(2,6)
circuit.cx(0,6)
circuit.p(pi/64,6)
circuit.cx(0,6)
circuit.cx(2,6)

circuit.cx(1,6)
circuit.cx(0,6)
circuit.p(pi/128,6)
circuit.cx(0,6)
circuit.cx(1,6)
circuit.draw(output = 'mpl')
```
